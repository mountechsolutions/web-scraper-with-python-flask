```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_confirmed = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv')
df_deaths = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv')
df_recovered = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv')
```

```python
#dropping the province/state column
df_deaths.drop(labels=['Province/State'], inplace= True, axis=1)
df_recovered.drop(labels=['Province/State'], inplace= True, axis=1)
df_confirmed.drop(labels=['Province/State'], inplace= True, axis=1)
```

```python
#grouping data countrywise
df_deaths_country=df_deaths.groupby(by=['Country/Region']).sum()
df_confirmed_country=df_confirmed.groupby(by=['Country/Region']).sum()
df_recovered_country=df_recovered.groupby(by=['Country/Region']).sum()
```

```python
pd.set_option('display.max_rows', None)
df_deaths_country
```

```python
type(df_deaths_country)
```

```python
#EXTRACTING DATA OF NEPAL
df_confirmed_Nepal=df_confirmed_country.loc[df_confirmed_country.index.isin(['Nepal'])]
df_deaths_Nepal=df_deaths_country.loc[df_deaths_country.index.isin(['Nepal'])]
df_recovered_Nepal=df_recovered_country.loc[df_recovered_country.index.isin(['Nepal'])]
df_recovered_Nepal
```

```python
# dropping the country , lat and long columns
df_confirmed_Nepal.drop(labels=['Lat','Long'], inplace= True, axis=1)
df_deaths_Nepal.drop(labels=['Lat','Long'], inplace= True, axis=1)
df_recovered_Nepal.drop(labels=['Lat','Long'], inplace= True, axis=1)
df_recovered_Nepal
```

```python
df_confirmed_Nepal_arr=df_confirmed_Nepal.to_numpy()
df_deaths_Nepal_arr=df_deaths_Nepal.to_numpy()
df_recovered_Nepal_arr=df_recovered_Nepal.to_numpy()
```

# Death Ratio

```python
#death ratio=(total death/total confirmed cases)
selected_columns = df_deaths_country[["Total_Deaths"]]
selected_col = df_confirmed_country[["Total_Confirmed"]]
df_ratio1 = selected_columns.copy()
df_ratio2 = selected_col.copy()
df_ratio=pd.concat([df_ratio1, df_ratio2], axis=1)
df_ratio.head()
```

```python
df_ratio['death_ratio'] = (df_ratio['Total_Deaths'] / df_ratio['Total_Confirmed'])
df_ratio['death_ratio'].head()
```

```python
df_ratio_sort=df_ratio.sort_values(by='death_ratio')[0:5]
df_ratio_sort
```

```python
#country with highest death ratio
df_ratio_sort=df_ratio.sort_values(by='death_ratio', ascending=False)[0:5]
df_ratio_sort
```

```python
df_ratio_sort_drop=df_ratio_sort['death_ratio']
ax = df_ratio_sort_drop.plot.bar(rot=0)
```
