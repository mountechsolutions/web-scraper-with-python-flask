```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.worldometers.info/coronavirus/"
req = requests.get(url)
soup = BeautifulSoup(req.text, "html.parser")
```

```python
results = soup.find(id='main_table_countries_today')
```

```python
results_row = results.find_all('tr')
```

```python
# results_row
```

```python
list1 = []
for tr in results_row:
    td= tr.find_all('td')
    row = [i.text for i in td]
    list1.append(row)
```

```python
# list1
```

```python
list2 =[]
for outer in list1:
    for inner in outer:
        print(inner.replace('\n', ''))
    list2.append(outer)    
```

```python
list2
for outer in list2:
    while('' in outer) : 
        outer.remove('') 
```

```python
# list2
```

```python
#few of the front row consist of continent data so removing continent data
i=0
list_countries=[]
for outer in list2:
    if (i>8):
        list_countries.append(outer)
    else:
        pass
    i=i+1
```

```python
# list_countries
```

```python
list_countries=list_countries[:214]
```

```python
countries = []
total_cases = []
new_cases = []
total_deaths = []
new_deaths = []
total_recovered = []
active_cases = []
critical = []
total_cases_per_mil_pop = []
total_deaths_per_mil_pop=[]
total_test=[]
test_per_mil_pop=[]
population=[]
```

```python
for item in list_countries:
    i=0
    for inner in item:
        if(i==1):
            countries.append(inner)
        if(i==2):
            total_cases.append(inner)
        if(i==3):
            new_cases.append(inner)
        if(i==4):
            total_deaths.append(inner)
        if(i==5):
            new_deaths.append(inner)
        if(i==6):
            total_recovered.append(inner)
        if(i==7):
            active_cases.append(inner)
        if(i==8):
            critical.append(inner)
        if(i==9):
            total_cases_per_mil_pop.append(inner)
        if(i==10):
            total_deaths_per_mil_pop.append(inner)
        if(i==11):
            total_test.append(inner)
        if(i==12):
            test_per_mil_pop.append(inner)
        if(i==13):
            population.append(inner)

        i=i+1
        
```

```python
print(len(countries)) 
print(len(total_cases ))
print(len(new_cases ))
print(len(total_deaths ))
print(len(new_deaths))
print(len(total_recovered ))
print(len(active_cases))
print(len(critical))
print(len(total_cases_per_mil_pop ))
print(len(total_deaths_per_mil_pop))
print(len(total_test))
print(len(test_per_mil_pop))
print(len(population))
```

```python
covid19_table = {
    "Country": countries,
    "Total Cases": total_cases,
    "New Cases": new_cases,
    "Total Deaths": total_deaths,
    "New Deaths": new_deaths,
    "Total Recovered": total_recovered,
    "Active Cases": active_cases,
    "critical": critical,
    "total_1M_ pop": total_cases_per_mil_pop,
     "Total death/1M pop":total_deaths_per_mil_pop,
     "Total test":total_test,
    "Test/1M pop":test_per_mil_pop,
    "Population":population
}
```
